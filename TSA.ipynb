```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns

data= pd.read_csv('/content/mroz (1).csv')
```

```python
print(data.info())
print(data.describe())
```

```python
missing_values =data.isna().sum()
print("Missing values:\n", missing_values)
```

```python
hours_data = data['hours'].dropna()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(hours_data, label='Hours Worked')
plt.title('Hours Worked Over Time')
plt.xlabel('Time Index')
plt.ylabel('Hours')
plt.legend()
plt.show()

# Decompose the time series (assuming monthly seasonality here; adjust as needed)
decomposition = seasonal_decompose(hours_data, period=12, model='additive')
decomposition.plot()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=hours_data)
plt.title('Box Plot of Hours Worked')
plt.xlabel('Hours')
plt.show()
```

```python
hushhrs_data = data['hushrs'].dropna()

# 1. Plot the time series data for 'hushhrs'
plt.figure(figsize=(10, 5))
plt.plot(hushhrs_data, label='Hush Hours')
plt.title('Hush Hours Over Time')
plt.xlabel('Time Index')
plt.ylabel('Hush Hours')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=hushhrs_data)
plt.title('Box Plot of Hush Hours')
plt.xlabel('Hush Hours')
plt.show()
```

```python
decomposition = seasonal_decompose(hushhrs_data, period=12, model='additive')
decomposition.plot()
plt.show()
```
