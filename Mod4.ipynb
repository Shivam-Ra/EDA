```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data= pd.read_csv('/content/mroz (1).csv')
```

```python
data = data[['hours', 'hushrs']].dropna()
```

```python
# Analyzing 'hushhrs'
print("\n1-D Analysis for 'hushhrs':")
hushhrs_mean = data['hushrs'].mean()
hushhrs_median = data['hushrs'].median()
hushhrs_std = data['hushrs'].std()
print(f"Mean: {hushhrs_mean}, Median: {hushhrs_median}, Standard Deviation: {hushhrs_std}")

```

```python
# Analyzing 'hours'
print("\n1-D Analysis for 'hours':")
hours_mean = data['hours'].mean()
hours_median = data['hours'].median()
hours_std = data['hours'].std()
print(f"Mean: {hours_mean}, Median: {hours_median}, Standard Deviation: {hours_std}")
```

```python
# Calculating Correlation between 'hours' and 'hushhrs'
correlation = data['hours'].corr(data['hushrs'])
print("\nCorrelation between 'hours' and 'hushhrs':", correlation)

# Covariance
covariance = data['hours'].cov(data['hushrs'])
print("Covariance between 'hours' and 'hushrs':", covariance)
```

```python
data['hours_bins'] = pd.cut(data['hours'], bins=5)
data['hushhrs_bins'] = pd.cut(data['hushrs'], bins=5)
contingency_table = pd.crosstab(data['hours_bins'], data['hushhrs_bins'])
print("\nContingency Table (hours_bins vs hushhrs_bins):\n", contingency_table)

# n-D Statistical Data Analysis
# Pairwise Correlation Matrix
correlation_matrix = data[['hours', 'hushrs']].corr()
print("\nPairwise Correlation Matrix:\n", correlation_matrix)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='hours', y='hushrs', data=data)
plt.title('Scatter Plot of Hours vs Hush Hours')
plt.xlabel('Hours')
plt.ylabel('Hush Hours')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(data.index, data['hours'], 'o', color='blue')
plt.title('Dot Plot for Hours')
plt.xlabel('Index')
plt.ylabel('Hours')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=data.index, y='hushrs', data=data, palette="viridis")
plt.title('Bar Plot of Hush Hours')
plt.xlabel('Index')
plt.ylabel('Hush Hours')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=data[['hours', 'hushrs']])
plt.title('Box Plot of Hours and Hush Hours')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
