```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
db = pd.read_csv('21BCE3727.csv')
db.describe();
db.shape
db.info()

```

```python
#data cleaning
db.dropna(inplace=True)

# Check for any remaining missing values
missing_values = db.isnull().sum()

# Remove duplicates
db.drop_duplicates(inplace=True)

# Display cleaned data and missing values check
print("Missing Values After Cleaning:")
print(missing_values)

print("\nCleaned Data (First 5 rows):")
print(db.head())
```

```python
#univariate analysis
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Distribution of 'wage' (Income/Salary)
sns.histplot(db['wage'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Wage')
axes[0, 0].set_xlabel('Wage')
axes[0, 0].set_ylabel('Frequency')

# 2. Distribution of 'educ' (Years of Education)
sns.histplot(db['educ'], bins=15, kde=False, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Years of Education')
axes[0, 1].set_xlabel('Years of Education')
axes[0, 1].set_ylabel('Frequency')

# 3. Distribution of 'age' (Age of Individuals)
sns.histplot(db['age'], bins=20, kde=False, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Age')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Frequency')

# 4. Distribution of 'kidslt6' (Number of Children Under 6)
sns.histplot(db['kidslt6'], discrete=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Children Under 6')
axes[1, 1].set_xlabel('Number of Children Under 6')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()


```

```python
# Set up figure layout for histograms and box plots
fig, axes = plt.subplots(4, 2, figsize=(14, 20))

# 1. Distribution of 'hours' (Hours Worked)
sns.histplot(db['hours'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Hours Worked')
axes[0, 0].set_xlabel('Hours Worked')
axes[0, 0].set_ylabel('Frequency')
sns.boxplot(x=db['hours'], ax=axes[0, 1])
axes[0, 1].set_title('Box Plot of Hours Worked')

# 2. Distribution of 'faminc' (Family Income)
sns.histplot(db['faminc'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Family Income')
axes[1, 0].set_xlabel('Family Income')
axes[1, 0].set_ylabel('Frequency')
sns.boxplot(x=db['faminc'], ax=axes[1, 1])
axes[1, 1].set_title('Box Plot of Family Income')

# 3. Distribution of 'repwage' (Reported Wage)
sns.histplot(db['repwage'], bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Distribution of Reported Wage')
axes[2, 0].set_xlabel('Reported Wage')
axes[2, 0].set_ylabel('Frequency')
sns.boxplot(x=db['repwage'], ax=axes[2, 1])
axes[2, 1].set_title('Box Plot of Reported Wage')

# 4. Distribution of 'exper' (Years of Experience)
sns.histplot(db['exper'], bins=30, kde=True, ax=axes[3, 0])
axes[3, 0].set_title('Distribution of Years of Experience')
axes[3, 0].set_xlabel('Years of Experience')
axes[3, 0].set_ylabel('Frequency')
sns.boxplot(x=db['exper'], ax=axes[3, 1])
axes[3, 1].set_title('Box Plot of Years of Experience')

plt.tight_layout()
plt.show()
```

```python
# Set up figure layout for all bivariate analysis plots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 1. Scatter Plot of 'wage' vs. 'hours' (Relationship between wage and hours worked)
sns.scatterplot(x=db['hours'], y=db['wage'], ax=axes[0, 0])
axes[0, 0].set_title('Scatter Plot: Wage vs Hours Worked')
axes[0, 0].set_xlabel('Hours Worked')
axes[0, 0].set_ylabel('Wage')

# 2. Box Plot of 'wage' by 'educ' (Wage distribution by years of education)
sns.boxplot(x=db['educ'], y=db['wage'], ax=axes[0, 1])
axes[0, 1].set_title('Box Plot: Wage by Years of Education')
axes[0, 1].set_xlabel('Years of Education')
axes[0, 1].set_ylabel('Wage')

# 3. Scatter Plot of 'faminc' vs. 'exper' (Relationship between family income and experience)
sns.scatterplot(x=db['exper'], y=db['faminc'], ax=axes[1, 0])
axes[1, 0].set_title('Scatter Plot: Family Income vs Experience')
axes[1, 0].set_xlabel('Years of Experience')
axes[1, 0].set_ylabel('Family Income')

# 4. Scatter Plot of 'wage' vs. 'age' (Relationship between wage and age)
sns.scatterplot(x=db['age'], y=db['wage'], ax=axes[1, 1])
axes[1, 1].set_title('Scatter Plot: Wage vs Age')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Wage')

plt.tight_layout()
plt.show()


```

```python


# 1. Pair Plot: Relationship between multiple variables
sns.pairplot(db[['wage', 'hours', 'faminc', 'educ', 'age']])
plt.title('Pair Plot of Wage, Hours Worked, Family Income, Education, and Age')
plt.show()


```

```python
# 2. 3D Scatter Plot: Wage vs Experience vs Hours Worked
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(db['wage'], db['exper'], db['hours'], c='b', marker='o')
ax.set_xlabel('Wage')
ax.set_ylabel('Years of Experience')
ax.set_zlabel('Hours Worked')
plt.title('3D Scatter Plot: Wage vs Experience vs Hours Worked')
plt.show()
```

```python
# 3. Correlation Heatmap: Multivariate relationship among key variables
plt.figure(figsize=(10, 8))
subset_corr = db[['wage', 'hours', 'faminc', 'educ', 'age', 'exper']].corr()
sns.heatmap(subset_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Key Variables')
plt.show()
```
